```python
import pandas as pd
```

```python
# ─────────────────────────────────────────
# 1. LOAD
# ─────────────────────────────────────────
df = pd.read_csv('saas_churn_dataset_raw.csv')
 
print("Shape:", df.shape)
print("\nNulls:\n", df.isnull().sum())
```

```python
# ─────────────────────────────────────────
# 2. FIX DATA TYPES
# ─────────────────────────────────────────
df['signup_date']        = pd.to_datetime(df['signup_date'])
df['last_activity_date'] = pd.to_datetime(df['last_activity_date'])
df['churn_date']         = pd.to_datetime(df['churn_date'])
 
df['mrr']                      = df['mrr'].astype(float)
df['effective_mrr']            = df['effective_mrr'].astype(float)
df['discount_pct']             = df['discount_pct'].astype(float)
df['logins_per_week']          = df['logins_per_week'].astype(float)
df['nps_score']                = df['nps_score'].astype(int)
df['features_used']            = df['features_used'].astype(int)
df['support_tickets_last_90d'] = df['support_tickets_last_90d'].astype(int)
df['tenure_months']            = df['tenure_months'].astype(int)
df['team_size']                = df['team_size'].astype(int)
df['churned']                  = df['churned'].astype(int)
```

```python
# ─────────────────────────────────────────
# 3. TRIM WHITESPACE ON STRING COLUMNS
# ─────────────────────────────────────────
str_cols = ['plan', 'contract_type', 'country', 'industry', 'referral_source']
for col in str_cols:
    df[col] = df[col].str.strip()
```

```python
# ─────────────────────────────────────────
# 4. DROP DUPLICATES
# ─────────────────────────────────────────
before = len(df)
df = df.drop_duplicates(subset='customer_id')
print(f"\nDuplicates removed: {before - len(df)}")
```

```python
# ─────────────────────────────────────────
# 5. FINAL CHECK
# ─────────────────────────────────────────
print("\n--- CLEANED DATASET ---")
print(f"Total rows: {len(df)}")
print(f"Shape     : {df.shape}")
print(f"\nDtypes:\n{df.dtypes}")
 
```

```python
# ─────────────────────────────────────────
# 6. EXPORT
# ─────────────────────────────────────────
df.to_csv('customer_churn.csv', index=False)
print("\nExported: customer_churn.csv")
```
